```python
import pandas as pd
import ast
```

```python
chennai_cars = pd.read_excel("Downloads/cars_data/chennai_cars.xlsx")
kolkata_cars = pd.read_excel("Downloads/cars_data/kolkata_cars.xlsx")
jaipur_cars = pd.read_excel("Downloads/cars_data/jaipur_cars.xlsx")
hyderabad_cars = pd.read_excel("Downloads/cars_data/hyderabad_cars.xlsx")
bangalore_cars = pd.read_excel("Downloads/cars_data/bangalore_cars.xlsx")
delhi_cars = pd.read_excel("Downloads/cars_data/delhi_cars.xlsx")
```

```python
# Load the provided Excel file to inspect its structure
#file_path = 'Downloads/cars_data/chennai_cars.xlsx'
#excel_data = pd.ExcelFile(file_path)

# Checking sheet names and loading data from the first sheet
#sheet_names = excel_data.sheet_names
#chennai_car_data = excel_data.parse(sheet_names[0])

# Displaying the first few rows of the data to understand its structure

all_cars = pd.concat([chennai_cars,kolkata_cars,jaipur_cars,hyderabad_cars,bangalore_cars,delhi_cars])
all_cars
```

```python

# Extract and expand key-value pairs from relevant columns ('new_car_overview', 'new_car_feature', 'new_car_specs')

# Initialize lists to hold extracted data for each category
overview_data, feature_data, specs_data = [], [], []

# Iterate over each row and process the dictionary columns
for _, row in all_cars.iterrows():
    # Parse each JSON-like string to a dictionary
    try:
        # Convert string representation of dictionaries to actual dictionaries
        overview = ast.literal_eval(row['new_car_overview'])
        feature = ast.literal_eval(row['new_car_feature'])
        specs = ast.literal_eval(row['new_car_specs'])
        
        # Extract key-value pairs from 'top' items if they exist in each section
        if 'top' in overview:
            overview_data.append({item['key']: item['value'] for item in overview['top']})
        else:
            overview_data.append({})

        if 'top' in feature:
            feature_data.append({item['value']: True for item in feature['top']})  # Use presence as indicator
        else:
            feature_data.append({})

        if 'top' in specs:
            specs_data.append({item['key']: item['value'] for item in specs['top']})
        else:
            specs_data.append({})

    except (ValueError, SyntaxError):
        # In case of any parsing errors, append empty dictionaries
        overview_data.append({})
        feature_data.append({})
        specs_data.append({})

# Create DataFrames for each category
overview_df = pd.DataFrame(overview_data)
feature_df = pd.DataFrame(feature_data)
specs_df = pd.DataFrame(specs_data)

# Concatenate all extracted data into a single DataFrame
cars_final_df = pd.concat([overview_df, feature_df, specs_df], axis=1)
cars_final_df
```

```python
cars_data = all_cars.copy()
```

```python
cars_data['new_car_detail_dict'] = cars_data['new_car_detail'].apply(ast.literal_eval)

cars_data['fuel_type'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('ft'))
cars_data['body_type'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('bt'))
cars_data['kilometers_driven'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('km'))
cars_data['OEM'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('oem'))
cars_data['model'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('model'))
cars_data['model_year'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('modelYear'))
cars_data['price'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('price'))
cars_data['Owner'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('owner'))
cars_data['Variant_name'] = cars_data['new_car_detail_dict'].apply(lambda x: x.get('variantName'))
```

```python
cars_data['name'] = cars_data['model']+' '+(cars_data['Variant_name'])
cars_data['name']
```

```python
cars_data_detail  = cars_data[['fuel_type','body_type','kilometers_driven','OEM','model','model_year','price','Owner','Variant_name','name']]
print(cars_data_detail)
```

```python
cars_data_details = pd.DataFrame(cars_data_detail)
cars_data_details
```

```python
final_data = cars_data_details.join(cars_final_df)
final_data
```
