```python
cars_data = df[['name','OEM','model_year','fuel_type','Transmission','Ownership','Mileage','Engine','Max Power','Torque','Seats','kilometers_driven','price']]
cars_data.head(4)
```

 # Data Preprocessing

```python
cars_data.dtypes
```

```python
cars_data['Engine'] = cars_data['Engine'].str.extract(r"(\d+)").astype(int)
```

```python
cars_data['Engine']
```

```python
# Extract numeric part, convert to float, and handle errors
cars_data['Mileage'] = pd.to_numeric(cars_data['Mileage'].str.extract(r"(\d+\.?\d*)")[0], errors='coerce')

# Replace NaNs with the mean of the column
cars_data['Mileage'].fillna(cars_data['Mileage'].mean(), inplace=True)

print("Data after handling non-numeric values and replacing nulls with mean:")
print(cars_data)
```

```python
cars_data.isnull().sum()
```

```python
# Extract numeric part, convert to float, and handle errors
cars_data['Max Power'] = pd.to_numeric(cars_data['Max Power'].str.extract(r"(\d+\.?\d*)")[0], errors='coerce')

# Replace NaNs with the mean of the column
cars_data['Max Power'].fillna(cars_data['Max Power'].mean(), inplace=True)

print("Data after handling non-numeric values and replacing nulls with mean:")
print(cars_data)
```

```python
cars_data.isnull().sum()
```

```python
import re

# Replace NaNs with '0nm' for consistent format
cars_data['Torque'] = cars_data['Torque'].fillna('0nm')

# Remove non-numeric characters and convert to integer
cars_data['Torque'] = cars_data['Torque'].apply(lambda x: int(re.findall(r'\d+', str(x))[0]) if re.findall(r'\d+', str(x)) else 0)

print(cars_data)
```

```python

# Ensure the 'Torque' column is treated as strings
cars_data['Torque'] = cars_data['Torque'].astype(str)

# Extract numeric part, convert to float, and handle errors
cars_data['Torque'] = pd.to_numeric(cars_data['Torque'].str.extract(r"(\d+\.?\d*)")[0], errors='coerce')

# Replace NaNs with the mean of the column
cars_data['Torque'].fillna(cars_data['Torque'].mean(), inplace=True)

print("Data after handling non-numeric values and replacing nulls with mean:")
print(cars_data)
```

```python
cars_data.dtypes
```

```python
# Extract numeric part, convert to float, and handle errors
cars_data['Seats'] = pd.to_numeric(cars_data['Seats'].str.extract(r"(\d+)")[0], errors='coerce')

# Replace NaNs with the mean of the column
cars_data['Seats'].fillna(cars_data['Seats'].mode()[0], inplace=True)

print("Data after handling non-numeric values and replacing nulls with mean:")
print(cars_data)
```

```python
cars_data.isnull().sum()
```

```python
mode_value= df['Ownership'].mode()[0]  # Get the mode of the column
df['Ownership'].fillna(mode_value, inplace=True)
```

```python
cars_data['Ownership'].fillna(cars_data['Ownership'].mode()[0], inplace=True)
```

```python
cars_data.isnull().sum()
```

```python
# Remove the "₹" symbol and strip spaces for each entry in the 'price' column
cars_data['price'] = cars_data['price'].str.replace("₹", "").str.strip()
```

```python
# Convert prices in Lakh, Million, or Crore to integer values in Rupees
def convert_price(price):
    match = re.match(r"(\d+\.\d+|\d+)\s*(Lakh|Million|Crore)?", price)
    if match:
        value = float(match.group(1))
        unit = match.group(2)
        if unit == "Lakh":
            value *= 100000
        elif unit == "Million":
            value *= 1000000
        elif unit == "Crore":
            value *= 10000000
        return int(value)
    return 0  # default if format doesn't match

cars_data['price'] = cars_data['price'].apply(convert_price)
print(cars_data)
```

```python
# Remove non-numeric characters and extra spaces, then convert to integer
cars_data['kilometers_driven'] = cars_data['kilometers_driven'].str.replace(r'\D', '', regex=True).astype(int)

print(cars_data.dtypes)  # Verify the new data type
print(cars_data)
```

```python
# Convert the 'Year' column to datetime format
cars_data['model_year'] = pd.to_datetime(cars_data['model_year'], format="%Y")

print(cars_data.dtypes)  # Verify the new data type
print(cars_data['model_year'].head())
```

```python
cars_data['Year'] = cars_data['model_year'].dt.year
```
